# Chat Scrapping from Youtube Live

```python
import csv
import pytchat
```

```python
#Copy paste your you tube link
youtube_link = "https://www.youtube.com/watch?v=yX1YuwOvSvY"
video_id = youtube_link.split("=")[1]
chat = pytchat.create(video_id=video_id)
```

```python
with open('youtube_chat.csv', 'a', newline='') as file:
    writer = csv.writer(file)
    writer.writerow(["Time", "Name", "Comment"])
    while chat.is_alive():
        for c in chat.get().sync_items():
            print(f"{c.datetime} [{c.author.name}]- {c.message}")
            writer.writerow([c.datetime, c.author.name, c.message])
            file.flush()
```
